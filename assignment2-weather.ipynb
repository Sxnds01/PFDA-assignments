# Plot of temperature over time using matplotlib
---
This notebook contains the dataset and the line plot of temperature of time.

```python
# First step is importing PANDAS
import pandas as pd
```

```python
# Then import the csv file containg the raw data
df = pd.read_csv('https://raw.githubusercontent.com/andrewbeattycourseware/PFDA-courseware/refs/heads/main/assignment/weatherreadings1.csv')
```

```python
df
```

```python
# Import matplotlib to plot the data
import matplotlib.pyplot as plt
```

```python
time = df['reportStartDateTime']
temperature = df['dryBulbTemperature_Celsius']

# Create a line plot
plt.plot(time, temperature, color = 'r', marker = '.')
plt.xlabel('Time')
plt.ylabel('Temperature in Celsius')
plt.title('Plot of Temperature over Time')

plt.grid(True)

# Show the plot
plt.show()
```
