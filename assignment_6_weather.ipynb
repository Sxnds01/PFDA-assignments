# Plot of Knock Airport Weather
---
This notebook contains the plots of different variables contained in the Knock Airport Weather dataset.

```python
# Import Pandas
import pandas as pd
import numpy as np
```

```python
# Read in the data using pandas but it skips the first few rows that were explaining the different headings
df = pd.read_csv('https://cli.fusio.net/cli/climate_data/webdata/hly4935.csv', skiprows= 23)
```

```python
df.head(10)
```

```python
# Change the date into the DateTime format
df['date'] = pd.to_datetime(df['date'], format='%d-%b-%Y %H:%M')

df.head(10)
```

```python
# Checking for gaps or missing values
df.iloc[157863]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Check what data types the columns are
print(df.dtypes)
```

```python
# Checking columns with object type to find invalid values
for col in df.select_dtypes(include='object').columns:
    print(f"Unique values in '{col}': {df[col].unique()[:10]}")  # Show first 10 unique values
```

```python
# Columns with spaces or special values to clean
columns_to_clean = ['rain', 'wddir', 'w', 'clamt', 'clht']

# Replace ' ' and 999 with NaN
df[columns_to_clean] = df[columns_to_clean].replace([' ', 999], np.nan)
```

```python
# Convert the object columns to numeric
df[columns_to_clean] = df[columns_to_clean].apply(pd.to_numeric, errors='coerce')
```

```python
# Replace NaN with 0
data = df.fillna(0)
```

```python
print(data.dtypes)  # Confirm all columns are numeric
print(data.isnull().sum())  # Confirm no NaN values 
```

```python
# List the columns that should be numeric
columns_to_convert = ['msl', 'wdsp', 'ww', 'vis']

# Convert columns to numeric (invalid values become NaN)
data[columns_to_convert] = data[columns_to_convert].apply(pd.to_numeric, errors='coerce')
```

```python
print(data.dtypes)
```

```python
data = data.fillna(0)
```

```python
data.iloc[145454]
```

```python
data.head(5)
```

```python
# Plot the temperatures from the new data frame

plt.figure(figsize=(10, 6))
plt.plot(data['date'], data['temp'], alpha=0.5)
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.title('The Temperature of Each Day')
plt.grid()
plt.show()
```

```python
# Get the mean temperature
data['day'] = data['date'].dt.date  # Extract only the date (no time)

grouped_data = data.groupby('day').mean()  # Calculate the mean for each day

grouped_data = grouped_data.reset_index()


```

```python
grouped_data.head(5)
```

```python
# Plot the mean temperatures

plt.figure(figsize=(10, 6))
plt.plot(grouped_data['date'], grouped_data['temp'], alpha=0.5) 
plt.xlabel('Date')
plt.ylabel('Mean Temperature')
plt.title('The Mean Temperature of Each Day')
plt.grid()
plt.show()
```

```python
# Plot the mena temperature of each month

grouped_data['month'] = pd.to_datetime(grouped_data['day']).dt.to_period('M')  # Extract month (YYYY-MM)

# Step 3: Calculate the monthly mean of the temperature
monthly_mean_temp = grouped_data.groupby('month')['temp'].mean()

# Step 4: Plot the monthly means
plt.figure(figsize=(10, 6))
plt.plot(monthly_mean_temp.index.astype(str), monthly_mean_temp.values)
plt.xlabel('Month')
plt.ylabel('Mean Temperature')
plt.title('Monthly Mean Temperature')
plt.grid()
plt.show()
```

```python
# Plot the windspeed

plt.figure(figsize=(10, 6))
plt.plot(data['date'], data['wdsp'], alpha=0.5)
plt.xlabel('Date')
plt.ylabel('Windspeed')
plt.title('Daily Windspeed')
plt.grid()
plt.show()
```

```python
data_wdsp = data 

data_wdsp.set_index('date', inplace=True)

data_wdsp.head(5)
```

```python
data_wdsp.drop(columns='level_0', inplace=True)
```

```python
data_wdsp.reset_index(inplace=True)
data_wdsp.head(5)
```

```python
# Calculate the rolling windspeed over a 24hr period

data_wdsp = data 

# Calculate rolling mean for wind speed over 24-hour period. Using a window of 24 since data is hourly.
data_wdsp['rolling_wdsp'] = data_wdsp['wdsp'].rolling(window=24).mean()


# Plot rolling wind speed
plt.figure(figsize=(10, 6))
plt.plot(data_wdsp['date'], data_wdsp['rolling_wdsp'], alpha=0.5)
plt.xlabel('Date')
plt.ylabel('Wind Speed (wdsp)')
plt.title('24-Hour Rolling Mean of Wind Speed')
plt.grid()
plt.show()

```

```python
data['day'] = data['date'].dt.date  # Extract only the date (no time)

daily_max_wind = data.groupby('day')['wdsp'].max()  # Calculate the max for each day

daily_max_wind = daily_max_wind.reset_index()

daily_max_wind.head(2)
```

```python
# Plot the daily max windspeed

plt.figure(figsize=(10, 6))
plt.plot(daily_max_wind['day'], daily_max_wind['wdsp'], alpha=0.5)
plt.xlabel('Date')
plt.ylabel('Wind Speed (wdsp)')
plt.title('Daily Maximum Wind Speed')
plt.grid()
plt.show()
```

```python
daily_max_wind['month'] = pd.to_datetime(daily_max_wind['day']).dt.to_period('M')  # Extract month (YYYY-MM)

# Step 3: Calculate the monthly mean of the daily max windspeeds
monthly_mean_wind = daily_max_wind.groupby('month')['wdsp'].mean()

# Step 4: Plot the monthly means
plt.figure(figsize=(10, 6))
plt.plot(monthly_mean_wind.index.astype(str), monthly_mean_wind.values)
plt.xlabel('Month')
plt.ylabel('Mean Windspeed (wdsp)')
plt.title('Monthly Mean of Daily Maximum Windspeeds')
plt.tight_layout()
plt.grid()
plt.show()
```
