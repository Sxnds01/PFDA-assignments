# Plot of Knock Airport Weather
---
This notebook contains the plots of different variables contained in the Knock Airport Weather dataset.

```python
# Import Pandas
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://cli.fusio.net/cli/climate_data/webdata/hly4935.csv', skiprows= 23)
```

```python
df.head(10)
```

```python
import pandas as pd

# Change the date into the DateTime format
df['date'] = pd.to_datetime(df['date'], format='%d-%b-%Y %H:%M')

df.head(10)
```

```python
# Checking for gaps or missing values
df.iloc[157863]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Check what data types the columns are
print(df.dtypes)
```

```python
# Checking columns with object type to find invalid values
for col in df.select_dtypes(include='object').columns:
    print(f"Unique values in '{col}': {df[col].unique()[:10]}")  # Show first 10 unique values
```

```python
# Columns with spaces or special values to clean
columns_to_clean = ['rain', 'wddir', 'w', 'clamt', 'clht']

# Replace ' ' and 999 with NaN
df[columns_to_clean] = df[columns_to_clean].replace([' ', 999], np.nan)
```

```python
# Convert the object columns to numeric
df[columns_to_clean] = df[columns_to_clean].apply(pd.to_numeric, errors='coerce')
```

```python
# Replace NaN with 0
data = df.fillna(0)
```

```python
print(data.dtypes)  # Confirm all columns are numeric
print(data.isnull().sum())  # Confirm no NaN values 
```

```python
# List the columns that should be numeric
columns_to_convert = ['msl', 'wdsp', 'ww', 'vis']

# Convert columns to numeric (invalid values become NaN)
data[columns_to_convert] = data[columns_to_convert].apply(pd.to_numeric, errors='coerce')
```

```python
print(data.dtypes)
```

```python
data = data.fillna(0)
```

```python
data.iloc[145454]
```

```python
data.head(5)
```

```python
# Plot the temperatures from the new data frame
import matplotlib.pyplot as plt

plt.plot(data['date'], data['temp'], alpha=0.5)
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.title('The Temperature of Each Day')
plt.show()
```

```python
# Get the mean temperature
data['day'] = data['date'].dt.date  # Extract only the date (no time)

grouped_data = data.groupby('day').mean()  # Calculate the mean for each day

grouped_data = grouped_data.reset_index()


```

```python
grouped_data.head(10)
```

```python
# Plot the mean temperatures
import matplotlib.pyplot as plt

plt.plot(grouped_data['date'], grouped_data['temp'], alpha=0.5) 
plt.xlabel('Date')
plt.ylabel('Mean Temperature')
plt.title('The Mean Temperature of Each Day')
plt.show()
```
