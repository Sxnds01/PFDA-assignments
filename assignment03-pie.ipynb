This notebook contains 1000 people's email domains, separated into three categories - @example.com, @example.net and @example.org. A pie chart is created from these categories with the percentage or occurance for each domain.

```python
import pandas as pd
import matplotlib.pyplot as plt

# creating dataframe
dataFrame = pd.DataFrame({
   "Email_domain": ['domain_com', 'domain_net', 'domain_org'],"Number_of_domains": [339, 320, 341]
})
colors = ['pink', 'steelblue', 'red']
explode = [0, 0, 0.1]

plt.pie(dataFrame["Number_of_domains"], labels = dataFrame["Email_domain"], autopct='%1.1f%%', startangle=90, explode = explode, shadow = True, colors = colors)

plt.title('Percentages of the Three different Email Domain Types')
plt.legend()
plt.show()
```
